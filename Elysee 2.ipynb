Taking a look at the public calendar of the French president François Hollande. The data is downloadable as a Json file from the presidency's website. 

Let's start by importing the data.

we have to say convert_dates=False because otherwise it converts the date column into a datetime format, but reads certain entries in a European format and others.

```python

from pandas import read_json, set_option, to_datetime

set_option('display.max_columns', 500)
calendar = read_json("http://www.elysee.fr/chronologie/download/json",convert_dates=False)
calendar
```

let's convert everything to the same date format

```python

calendar["date"]=to_datetime(calendar["date"], format="%d/%m/%Y")
```

it works!


```python
(calendar["date"][999])
```

Let's look at what he did the most


```python
calendar["title"].value_counts()
```

The problem with this is that we assume that the entries in the "title" column are all in the same format, which seems unlikely, given the fact that the top numbers are pretty low. Using regular expressions should lead to a better result. 

Let's look at where he holds his meetings. 
maybe we can map this and somehow calculate the cost of his travels. 

```python
calendar["place"].value_counts()
```

The "Type" column tells us if the meeting was in the Elysee palace, somewhere else in France or abroad. 

```python
calendar["type"].value_counts()
```

```python
calendar[calendar["type"]=="elysee"].shape[0]
```

Let's calculate the percentages 

```python
france_count = float(calendar[calendar["type"]=="elysee"].shape[0]+calendar[calendar["type"]=="france"].shape[0])/float(sum(calendar["type"].value_counts()))
```

```python
international_count = float(calendar[calendar["type"]=="international"].shape[0])/float(sum(calendar["type"].value_counts()))
```

```python
france_count+international_count
```
